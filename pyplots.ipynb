```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro']
```

```python
with open('data/count_distractors_all.tsv', 'r') as f:
    lines = [line.strip().split('\t') for line in f]
ls1 = [''.join(line[0].split(' ')) for line in lines]
y1 = [int(line[1]) for line in lines]

with open('data/count_distractors_baseline.tsv', 'r') as f:
    lines = [line.strip().split('\t') for line in f]
ls2 = [''.join(line[0].split(' ')) for line in lines]
y2 = [int(line[1]) for line in lines]

with open('data/count_distractors_baseline_previous.tsv', 'r') as f:
    lines = [line.strip().split('\t') for line in f]
ls3 = [''.join(line[0].split(' ')) for line in lines]
y3 = [int(line[1]) for line in lines]
```

```python
# 各問題の上位75頻度のdistractor

length = 75
# w = 0.7

# fig = plt.figure(figsize=(10.0, length))
# x = np.array([n * 1.5 for n in range(length)])
# ls = [f'{l1}\n\n{l2}' for l1, l2 in zip(ls1[:length][::-1], ls2[:length][::-1])]
# plt.barh(x+w, y1[:length][::-1], height=w)
# plt.barh(x, y2[:length][::-1], height=w, color='crimson')
# plt.yticks(x+w/2, ls)
# plt.xlim(0, max(y1+y2) + 25)
# plt.grid(True)
# plt.show()
```

```python
# 全条件を適用して作成した問題の上位75頻度のdistractor
# 「問題」「買う」「購入」「食べる」「利用」といった単語が多く含まれ、偏りがある印象
# Length条件はどちらも除いていないのに、文長が短い傾向にあるように思える

fig = plt.figure(figsize=(10.0, 25))
plt.barh(ls1[:length][::-1], y1[:length][::-1], height=1)
plt.xlim(0, 250)
plt.grid(True)
plt.show()
```

```python
# Co-occurとDifficultを除いて作成した問題の上位75頻度のdistractor

fig = plt.figure(figsize=(10.0, 25))
plt.barh(ls2[:length][::-1], y2[:length][::-1], height=1, color='crimson')
plt.xlim(0, 250)
plt.grid(True)
plt.show()
```

```python
# 誤り選択肢の抽出時に、類似度の最も高いもの、中央値を取るもの、最も低いものを取得するという条件でやった場合

fig = plt.figure(figsize=(10.0, 25))
plt.barh(ls3[:length][::-1], y3[:length][::-1], height=1, color='green')
plt.xlim(0, 3000)
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(10.0, 25))
plt.barh(ls3[:length][::-1], y3[:length][::-1], height=1, color='green')
plt.xlim(0, 250)
plt.grid(True)
plt.show()
```
